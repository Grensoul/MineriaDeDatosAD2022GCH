# Anova

```python
from typing import Tuple, List
import re
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
df.info()
```

```python
def anova(df_aux: pd.DataFrame, str_ols: str):
    modl = ols(str_ols, data=df_aux).fit()
    anova_df = sm.stats.anova_lm(modl, typ=2)
    
    if anova_df["PR(>F)"][0] < 0.005:
        print("hay diferencias")
        print(anova_df)
    else:
        print("No hay diferencias")

def anova_1(df_complete: pd.DataFrame):
    df_by_type = df_complete.groupby(["Currency", 
                                      "Released"])[["Price"]].aggregate(pd.DataFrame.sum)
    df_by_type.reset_index(inplace=True)
    df_by_type.set_index("Released", inplace=True)
    df_by_type.reset_index(inplace=True)
    
    
    df_aux = df_by_type.rename(columns={"Price": "Gasto"}).drop(['Released'], axis=1)
    print(df_aux.head())
    anova(df_aux, "Gasto ~ Currency")
```

```python
anova_1(df)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
# print the coeficients
print(lm.coef_)
```

```python
# print the intercept
print(lm.intercept_)
```

```python
coeff_df = pd.DataFrame(lm.coef_,X.columns,columns=['Rating Coefficient'])
coeff_df
```

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```

```python
np.reshape(lm.coef_,4)[0]
```

```python
np.reshape(lm.intercept_,1)
```

```python
def abline(coef,intercept):
    x = np.linspace(-5,5,100)
    y = coef*x+intercept  #y=ax+b
    
    plt.plot(x, y, '-r', label='Coef[0] (Rating trending) y='+str(coef)+'x+'+str(intercept))
    plt.title('Graph of y='+str(coef)+'x+'+str(intercept))
    plt.xlabel('Installs', color='#1C2833')
    plt.ylabel('Rating', color='#1C2833')
    plt.grid()
    plt.show()
    
abline(np.reshape(lm.coef_,4)[0],np.reshape(lm.intercept_,1))
```
